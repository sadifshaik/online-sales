```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('online_sales_dataset.csv')
data.head(10)
```

```python
data.describe()
```

```python
plt.hist(data['UnitPrice'])
plt.xlabel('Unit Price')
plt.xlabel('Unit Price')
plt.title('Histogram of Unit price for stocks')
plt.show()
```

```python
plt.hist(data['Discount'])
plt.ylabel('Stocks')
plt.xlabel('Discount')
plt.title('Histogram for stocks and discount')
plt.show()
```

```python
plt.bar(data['Quantity'],data['ShippingCost'], width=0.5, color='red')
plt.xlabel('Quantity of Stock')
plt.ylabel('Shipping cost')
plt.title('Quantity vs Shipping cost')
plt.show()
```

```python
plt.plot(data['Discount'], color='green')
plt.xlabel('Stocks')
plt.title('line plot of Discount')
plt.ylabel('Discount')
plt.show()
```

```python
sns.boxplot(data['ShippingCost'])
```
